```python
import yaml
import pandas as pd

def read_credentials(credential_file_path):
    with open(credential_file_path) as f:
        cfg = yaml.load(f, Loader=yaml.FullLoader)
    return cfg


config = read_credentials('credentials.yaml')
config
```

```python
# IMPORT THE SQALCHEMY LIBRARY's CREATE_ENGINE METHOD
from sqlalchemy import create_engine
 
# Connection details
hostname = config.get("hostname", "")
database = config.get("database", "")
port = config.get("port", "")
username = config.get("username", "")
password = config.get("password", "")


schema_name = config.get("schema_name", "")
table_name = config.get("table_name", "")

 
# PYTHON FUNCTION TO CONNECT TO THE MYSQL DATABASE AND
# RETURN THE SQLACHEMY ENGINE OBJECT
def get_connection():

    connection_string = f"postgresql+psycopg2://{username}:{password}@{hostname}:{port}/{database}"
    return create_engine(connection_string)
```

```python
engine  = get_connection()
```

```python
schema_name = "myschema"
table_name = "machine_data"
```

```python
# Read the table into a pandas DataFrame
query = f'SELECT * FROM {schema_name}.{table_name}'
```

```python
df = pd.read_sql(query, con=engine)
```
